```python
#this will generate the probabilities of each word
def UnigramModel(training_set, total_words, probabilites):
  probabilities = dict()
  i = 0

  while(i < total_words):                               # in a single loop simply count occurences of each word
    if str(training_set[i]) in probabilities:
      probabilities[str(training_set[i])] += 1
    else:
      probabilities[str(training_set[i])] = 1
    i += 1

  print(len(probabilities))                             #will tell us how many unique characters exist
  #for prob in probabilities :
  #  print(prob, end = "  -> ")
  #  print(probabilities[prob])
```

```python
#Function to build our simple bigram models which will return the word most probable to come after given word
def BigramModel(word, training_set, total_words):
  store_occurences = dict()
  i = 0
  
  while(i < total_words-1):
    if(str(training_set[i]) == word):
      if str(training_set[i+1]) in store_occurences:
        store_occurences[str(training_set[i+1])] += 1
      else :
        store_occurences[str(training_set[i+1])] = 1
    i += 1

  for key, value in sorted(store_occurences.items(), key=lambda item: item[1], reverse = True):
    #print(key, end = "  -> ")
    #print(value)
    if(key != "-" and key != "‘" and key != "%" and key != "٪" and key != "“" and key != "’" and key != "!" and key != "؟" and key != "،" and key != "."):
      return key
```

```python
#Function to build our trigram models
def TrigramModel(word1,word2 ,training_set, total_words):
  store_occurences = dict()
  i = 0
  while(i < total_words-2):
    if(str(training_set[i]) == word1):
      if(str(training_set[i+1]) == word2):
        if str(training_set[i+2]) in store_occurences:
          store_occurences[str(training_set[i+2])] += 1
        else :
          store_occurences[str(training_set[i+2])] = 1
    i += 1

  for key, value in sorted(store_occurences.items(), key=lambda item: item[1], reverse = True):
    #print(key, end = "  -> ")
    #print(value)
    if(key != "-" and key != "‘" and key != "%" and key != "٪" and key != "“" and key != "’" and key != "!" and key != "؟" and key != "،" and key != "."):
      return key
```

```python
#program to build bigram model which will help generate poetry starting from last word first i.e left to right
def BackwardBigramModel(word, training_set, total_words):
  store_occurences = dict()
  i = 1
  while(i < total_words):
    if(str(training_set[i]) == word):
      if str(training_set[i-1]) in store_occurences:
        store_occurences[str(training_set[i-1])] += 1
      else :
        store_occurences[str(training_set[i-1])] = 1
    i += 1

  for key, value in sorted(store_occurences.items(), key=lambda item: item[1], reverse = True):
    #print(key, end = "  -> ")
    #print(value)
    if(key != "-" and key != "‘" and key != "%" and key != "٪" and key != "“" and key != "’" and key != "!" and key != "؟" and key != "،" and key != "."):
      return key
```

```python
import random #for random number generation b/w 7-10 and for randomly choosing words from training set

#generate poem using BiGram Model
def generatepoem_usingBiGram(training_set, starting_words, total_words):
  word = ""
  poem = ""
  word_to_rhyme = ""

  for k in range(1,4):
    for x in range(1,5):
      words_in_stanza = random.randint(7,10)
      for y in range(0, words_in_stanza):
        if(y == 0):
          #word = start[k-1][x-1]                                 # -> IMP: UNCOMMENT THIS LINE AND COMMENT THE NEXT 3 TO COMPARE RESULTS OF BIGRAM AND TRIGRAM IN A CONTROLLED ENVIRONMENT
          word = str(random.choice(starting_words)) 
          while(word == "-" or word == "‘" or word == "%" or word == "٪" or word == "“" or word == "’" or word == "!" or word == "؟" or word == "،" or word == "."):
            word = str(random.choice(starting_words))
        else:
          if y == words_in_stanza-1 and (x == 2 or x == 4):       #i am trying to make the 2nd and 4th stanzas rhyme 
            word = rhymingwords(word_to_rhyme,training_set, total_words)
            if(word == None):
              word = BigramModel(word, training_set, total_words)
              if word == None:
                word = str(random.choice(training_set))
                while(word == "-" or word == "‘" or word == "%" or word == "٪" or word == "“" or word == "’" or word == "!" or word == "؟" or word == "،" or word == "."):
                  word = str(random.choice(training_set))
          else:
            word = BigramModel(word, training_set, total_words)
            if word == None:
              word = str(random.choice(training_set))
              while(word == "-" or word == "‘" or word == "%" or word == "٪" or word == "“" or word == "’" or word == "!" or word == "؟" or word == "،" or word == "."):
                word = str(random.choice(training_set))
        if y == words_in_stanza-1 and x == 2 and k == 1:          #the first time a second stanza comes along we will simply store the word to later rhyme with
          word_to_rhyme = word
        poem += word + " "
      poem += "\n"
    poem += "\n"

  return poem
```

```python
bigram_poem = generatepoem_usingBiGram(training_set, starting_words, total_words)
```

```python
#generate format for the poetry verses
def generatepoem_usingBackwardBiGram(training_set, starting_words, total_words):
  word = ""
  poem = ""
  word_to_rhyme = ""

  for k in range(1,4):
    for x in range(1,5):
      store_poem = list()
      words_in_stanza = random.randint(7,10)
      for y in range(0, words_in_stanza):
        if (y == 0):
          if (x ==2 or x == 4):
            word = rhymingwords(word_to_rhyme,training_set,total_words)
            if (word == None):    
              word = BackwardBigramModel(word,training_set, total_words) 
              if word == None:
                word = str(random.choice(training_set)) 
                while(word == "-" or word == "‘" or word == "%" or word == "٪" or word == "“" or word == "’" or word == "!" or word == "؟" or word == "،" or word == "."):
                  word = str(random.choice(training_set))
          else:
            word = str(random.choice(training_set)) 
            while(word == "-" or word == "‘" or word == "%" or word == "٪" or word == "“" or word == "’" or word == "!" or word == "؟" or word == "،" or word == "."):
              word = str(random.choice(training_set))
        else:
          word = BackwardBigramModel(word,training_set, total_words) 
          if word == None:
            word = str(random.choice(training_set)) 
            while(word == "-" or word == "‘" or word == "%" or word == "٪" or word == "“" or word == "’" or word == "!" or word == "؟" or word == "،" or word == "."):
              word = str(random.choice(training_set))
        store_poem.insert(0,word)
        if (y == 0 and x == 2 and k == 1):
          word_to_rhyme = word
      for eachwordinstorepoem in store_poem:
        poem = poem + eachwordinstorepoem + " "
      poem = poem + "\n"
    poem = poem + "\n"

  return poem
```

```python
#The idea is to randomly choose a word in the middle and then build on both side of the word using backward and normal bigram model simultaneously
def generatepoem_usingBidirectionalBigram(training_set, total_words):
  word1 = ""
  word2 = ""
  word = ""
  finalpoem = ""

  for k in range(1,4):
    for x in range(1,5):
      words_in_stanza = random.randint(7,10)
      words_in_stanza += (words_in_stanza % 2 > 0)
      ind = int(words_in_stanza/2)
      temp_stanza_store = [''] * words_in_stanza
      i = 1
      for y in range(ind, words_in_stanza-1):
        if(y == ind):
          word = str(random.choice(training_set)) 
          while(word == "-" or word == "‘" or word == "%" or word == "٪" or word == "“" or word == "’" or word == "!" or word == "؟" or word == "،" or word == "."):
            word = str(random.choice(training_set))
          temp_stanza_store[y] = word
          word1 = BigramModel(word, training_set, total_words)
          if word1 == None:
            word1 = str(random.choice(training_set))
            while(word1 == "-" or word1 == "‘" or word1 == "%" or word1 == "٪" or word1 == "“" or word1 == "’" or word1 == "!" or word1 == "؟" or word1 == "،" or word1 == "."):
              word1 = str(random.choice(training_set))
          word2 = BackwardBigramModel(word,training_set,total_words)
          if word2 == None:
            word2 = str(random.choice(training_set))
          temp_stanza_store[y+1] = word1
          temp_stanza_store[y-i] = word2
          i += 2
        else:
          word1 = BigramModel(word1, training_set, total_words)
          if word1 == None:
            word1 = str(random.choice(training_set))
            while(word1 == "-" or word1 == "‘" or word1 == "%" or word1 == "٪" or word1 == "“" or word1 == "’" or word1 == "!" or word1 == "؟" or word1 == "،" or word1 == "."):
              word1 = str(random.choice(training_set))
          word2 = BackwardBigramModel(word2, training_set, total_words)
          if word2 == None:
            word2 = str(random.choice(training_set))
            while(word2 == "-" or word2 == "‘" or word2 == "%" or word2 == "٪" or word2 == "“" or word2 == "’" or word2 == "!" or word2 == "؟" or word2 == "،" or word2 == "."):
              word2 = str(random.choice(training_set))
          temp_stanza_store[y+1] = word1
          temp_stanza_store[y-i] = word2
          i += 2
     #   print(temp_stanza_store)
     # print()
      count = len(temp_stanza_store) - 1
      while(count >= 0):
        finalpoem += temp_stanza_store[count] + " "
        count -= 1
      finalpoem = finalpoem + "\n"
    #print()
    finalpoem = finalpoem + "\n"

  return finalpoem
```
